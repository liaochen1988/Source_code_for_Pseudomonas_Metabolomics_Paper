```python
import BBH_python3
import pandas as pd
import os
```

# We only include strains that have rhamnnolipid production data

```python
df_rhl = pd.read_excel('../data/rhamnolipids/rhamnMat.xlsx', index_col=0)
df_rhl.head()
```

# Do reciprocal blast

```python
subject = 'PA_UCBPP-PA14_feature_protein'

for strain in df_rhl.index:
    if strain == 'PA14':
        continue
    outfile = 'PA14_vs_'+strain+'.txt'
    if not os.path.isfile(outfile):
        query = 'PA_' + strain + '_feature_protein'
        BBH_python3.make_blast_db(subject)
        BBH_python3.make_blast_db(query)
        BBH_python3.get_bbh(query, subject, in_folder='bbh', outfile=outfile)
```

# Find correspondence between fig proteins

```python
df_orth = pd.DataFrame()
for index,strain in enumerate(df_rhl.index):
    if strain=='PA14':
        continue
    infile = 'PA14_vs_'+strain+'.txt'
    df = pd.read_csv(infile, sep=',', index_col=0).reset_index()
    df = df[df.BBH == '<=>'] # select only orthologue
    df = df[['gene','subject']]
    df.columns = [strain, 'PA14']
    df = df.set_index('PA14')
    if index==0:
        df_orth = df
    else:
        df_orth = pd.merge(df_orth, df, left_index=True, right_index=True, how='outer')
        
df_orth.index = [x.rstrip('|') for x in df_orth.index]
df_orth.head()
```

```python
len(df_orth.index)
```

```python
#  add PATRIC annotation
df_PA14_feature = pd.read_csv('../PATRIC_GMM/UCBPP-PA14/PATRIC_genome_feature.txt', sep='\t')
df_PA14_feature = df_PA14_feature[['Product','PATRIC ID',]].set_index('PATRIC ID')
df_orth = pd.merge(df_orth, df_PA14_feature, left_index=True, right_index=True, how='left')
df_orth = df_orth[['Product']+[x for x in df_orth.columns if x != 'Product']]
df_orth.head()
```

```python
df_orth.to_csv('protein_orthologue_dictionary_ref_PA14_all.csv')
```
