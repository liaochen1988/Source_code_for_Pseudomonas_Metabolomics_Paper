```python
import pandas as pd
import os
import numpy as np
```

```python
df_rhl = pd.read_excel('../data/rhamnolipids/rhamnMat.xlsx', index_col=0)
df_rhl = df_rhl.rename(index={'PA14':'UCBPP-PA14'})
df_rhl.head()
```

```python
df_orth_plfam = None
for s in df_rhl.index:
    df_feature = pd.read_csv('../PATRIC_GMM/%s/PATRIC_genome_feature.txt'%(s), sep='\t')
    df_feature = df_feature[['PATRIC genus-specific families (PLfams)','Product','PATRIC ID',]]
    df_feature = df_feature.rename(columns={'PATRIC ID':s})
    df_feature = df_feature.dropna(axis=0, how='any')
    df_feature = df_feature.groupby(['PATRIC genus-specific families (PLfams)', 'Product'])[s].apply(tuple).to_frame().reset_index()
    df_feature[s] = [(',').join(x) for x in df_feature[s]]
    if df_orth_plfam is None:
        df_orth_plfam = df_feature
    else:
        df_orth_plfam = pd.merge(df_orth_plfam, 
                                 df_feature, 
                                 left_on=['PATRIC genus-specific families (PLfams)','Product'],
                                 right_on=['PATRIC genus-specific families (PLfams)','Product'],
                                 how='outer')
df_orth_plfam
```

```python
df_orth_plfam.to_csv('PLfam_dictionary.csv')
```

```python
df_orth_pgfam = None
for s in df_rhl.index:
    df_feature = pd.read_csv('../PATRIC_GMM/%s/PATRIC_genome_feature.txt'%(s), sep='\t')
    df_feature = df_feature[['PATRIC cross-genus families (PGfams)','Product','PATRIC ID',]]
    df_feature = df_feature.rename(columns={'PATRIC ID':s})
    df_feature = df_feature.dropna(axis=0, how='any')
    df_feature = df_feature.groupby(['PATRIC cross-genus families (PGfams)', 'Product'])[s].apply(tuple).to_frame().reset_index()
    df_feature[s] = [(',').join(x) for x in df_feature[s]]
    if df_orth_pgfam is None:
        df_orth_pgfam = df_feature
    else:
        df_orth_pgfam = pd.merge(df_orth_pgfam, 
                                 df_feature, 
                                 left_on=['PATRIC cross-genus families (PGfams)','Product'],
                                 right_on=['PATRIC cross-genus families (PGfams)','Product'],
                                 how='outer')
df_orth_pgfam
```

```python
df_orth_pgfam.to_csv('PGfam_dictionary.csv')
```
