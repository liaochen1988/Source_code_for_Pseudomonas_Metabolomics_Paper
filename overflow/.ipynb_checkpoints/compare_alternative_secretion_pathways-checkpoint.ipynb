# Customized functions

### set glycerol minimal medium

```python
def set_glycerol_minimal_medium(model, f_glyc, f_nh4, f_fe3):
    
    # set lower bound of all exchange reactions to zero
    for rxn in model.exchanges:
        rxn.lower_bound = 0

    # turn on specific fluxes
    minimal__media = {
            'EX_glyc_e_': -f_glyc,   # glycerol
            'EX_nh4_e_': -f_nh4,     # ammonium
            'EX_pi_e_':  -1000,      # phosphate
            'EX_so4_e_': -1000,      # sulfate
            'EX_fe3_e_': -f_fe3,     # fe3+
            'EX_mn2_e_': -1000,      # mn2+
            'EX_zn2_e_': -1000,      # zn2+
            'EX_cu2_e_': -1000,      # cu2+
            'EX_ca2_e_': -1000,      # ca2+
            'EX_cl_e_' : -1000,      # cl-
            'EX_cobalt2_e_': -1000,  # cobalt
            'EX_k_e_':   -1000,      # k
            'EX_mobd_e_'   : -1000,  # mo6+
            'EX_na1_e_'    : -1000,  # na+
            'EX_ni2_e_'    : -1000,  # ni2+
            'EX_mg2_e_': -1000,      # mg2+
            'EX_o2_e_':  -1000
        }

    for r in minimal__media.keys():
        try:
            rxn = model.reactions.get_by_id(r)
            rxn.lower_bound = minimal__media[r]
            rxn.upper_bound = 1000
        except:
            print("The reaction {} does not exist in model.".format(str(r)))
            continue
        
    return model
```

### add constraint to the total flux of a metabaolite

```python
def add_flux_constraint(model, compound_id):
    
    # find stoichiometric coefficients of all reactions that produce/consume a metabolite
    S_dict = {}
    for r in model.reactions:
        for m in r.reactants:
            if m.id == compound_id:
                S_dict[r.id] = r.get_coefficient(compound_id)
                break           
        for m in r.products:
            if m.id == compound_id:
                S_dict[r.id] = r.get_coefficient(compound_id)
                break
                    
    # add flux constraint
    producing_fluxes = []
    for k,v in S_dict.items():
        if v>0:
            varX = model.problem.Variable('X_' + k, lb=0, ub=1, type='binary')
            con0A = model.problem.Constraint(model.reactions.get_by_id(k).flux_expression,
                                             name = 'con0A_' + k,
                                             indicator_variable=varX,
                                             active_when=0,
                                             ub=0)
            con0B = model.problem.Constraint(model.reactions.get_by_id(k).flux_expression,
                                             name = 'con0B_' + k,
                                             indicator_variable=varX,
                                             active_when=1,
                                             lb=0)
        else:
            varX = model.problem.Variable('X_' + k, lb=0, ub=1, type='binary')
            con0A = model.problem.Constraint(model.reactions.get_by_id(k).flux_expression,
                                             name = 'con0A_' + k,
                                             indicator_variable=varX,
                                             active_when=0,
                                             lb=0)
            con0B = model.problem.Constraint(model.reactions.get_by_id(k).flux_expression,
                                             name = 'con0B_' + k,
                                             indicator_variable=varX,
                                             active_when=1,
                                             ub=0)

        # refer to https://www.researchgate.net/post/How_can_I_linearize_the_product_of_two_variables for how to encode Z = X * Flux
        flux_ub = model.reactions.get_by_id(k).upper_bound
        flux_lb = model.reactions.get_by_id(k).lower_bound
            
        varZ = model.problem.Variable('Z_' + k, lb=min([0,flux_lb]), ub=flux_ub)
        con1 = model.problem.Constraint(varZ - flux_ub*varX, name = 'con1_' + k, ub=0)
        con2 = model.problem.Constraint(varZ - flux_lb*varX, name = 'con2_' + k, lb=0)
        con3 = model.problem.Constraint(varZ - model.reactions.get_by_id(k).flux_expression + (1-varX)*flux_ub, name = 'con3_' + k, lb=0)
        con4 = model.problem.Constraint(varZ - model.reactions.get_by_id(k).flux_expression + (1-varX)*flux_lb, name = 'con4_' + k, ub=0)
                    
        producing_fluxes.append(varZ*v)
        model.add_cons_vars([varX, varZ, con0A, con0B, con1, con2, con3, con4])
        model.solver.update()   
        #print(k, v, model.slim_optimize())
            
    flux_constraint = model.problem.Constraint(add(*producing_fluxes), lb=-1000, ub=1000, name="flux_constraint__%s"%(compound_id))
    model.add_cons_vars([flux_constraint])
    model.solver.update()
    
    return model, S_dict
```

### set CPLEX parameters

```python
def set_cplex_parameters(model, verbose, compound_id=''):
    model.solver = 'cplex'
    model.solver.configuration.lp_method = "barrier"
    model.solver.configuration.qp_method = "barrier"
    model.solver.configuration.presolve = True
    model.solver.configuration.tolerances.integrality = 1e-6
    model.solver.problem.parameters.mip.tolerances.integrality.set(1e-6)
    model.solver.configuration.verbosity=verbose
    
    if compound_id in ['nadph_c']:
        model.solver.problem.parameters.emphasis.mip.set(3)
    else:
        model.solver.problem.parameters.emphasis.mip.set(1)
        
    model.solver.problem.parameters.mip.strategy.probe.set(3)
    model.solver.problem.parameters.mip.strategy.rinsheur.set(20)
    model.solver.problem.parameters.mip.strategy.nodeselect.set(2)
    model.solver.problem.parameters.mip.strategy.heuristicfreq.set(20)
    #model.solver.configuration.timeout = 30 # 30 s
    
    return model
```

### fix loops in NADH production and consumption

```python
def fix_NADH_NADPH_flux_loop(model):
    
    # NADTRHD, FMNRx and FMNR2r form a loop
    # NADTRHD: nad_c + nadph_c --> nadh_c + nadp_c
    # FMNRx: fmn_c + h_c + nadh_c --> fmnh2_c + nad_c        # missing in iMO1053
    # FMNR2r: fmn_c + h_c + nadph_c <=> fmnh2_c + nadp_c     # missing in iMO1053
    model.reactions.FMNR2r.lower_bound = 0
    
    # GLYD: h_c + hpyr_c + nadh_c <==> glyc_R_c + nad_c      # found in iMO1053, found in Pseudocyc (forward), found in Metacyc (bidirectional)
    # TRSARr: 2h3oppan_c + h_c + nadh_c <=> glyc_R_c + nad_c # missing in iMO1053,  missing in Pseudocyc, found in Metacyc (forward)
    model.reactions.GLYD.lower_bound = 0
    model.reactions.TRSARr.lower_bound = 0

    # Set of reactions that couple NADH/NAD to NADPH/NADP
    # the first two reactions are found in Metacyc, Pathway: 9-cis, 11-trans-octadecadienoyl-CoA degradation 
    # HACD31i: 3hhd58coa_c + nad_c --> 3ohd58ccoa_c + h_c + nadh_c
    # ECOAH36: 3hhd58coa_c <=> h2o_c + td58_2_coa_c                       # also found in Pseudocyc (Reverse)
    # RECOAH24: R_3htd58coa_c <=> h2o_c + td58_2_coa_c
    # RHACOAR142: 3ohd58coa_c + h_c + nadph_c <=> R_3htd58coa_c + nadp_c
    #
    # Reactions starting with "RECOAH" seems to be specific to PP but lack evidence in PA
    # We assume that they can flow in the reverse direction, similar to ECOAH36
#     for r in model.reactions:
#         if r.id.startswith('RECOAH'):
#             r.upper_bound = 0

    # We did not find evidences that NAD/NADH is substrate of fdxr_42_c in both Pseudocyc and iMO1053
    model.reactions.FRDO6r.lower_bound=0
    #model.reactions.FRDO6r.upper_bound=0
    model.reactions.FRDO7r.lower_bound=0
    #model.reactions.FRDO7r.upper_bound=0

    # GLUDx and GLUDy in Pseudocyc goes forward direction
    # GLUDx: glu__L_c + h2o_c + nad_c ⇌ akg_c + h_c + nadh_c + nh4_c
    # GLUDy: glu__L_c + h2o_c + nadp_c ⇌ akg_c + h_c + nadph_c + nh4_c
    model.reactions.GLUDx.lower_bound = 0
    #model.reactions.GLUDy.lower_bound = 0

    # CO2 will be gone after released
    #model.reactions.H2CO3D.upper_bound = 0
    #model.reactions.HCO3E.upper_bound = 0

    # SHK3Dr and SHK3D are the same reaction but conflict in reversibility
    #model.reactions.SHK3Dr.lower_bound = 0

    # PRPP is irreversible
    #model.reactions.PRPPS.lower_bound = 0
    
    # LEUDHr, VALDHr, ILEDHr are all reversible, which can be coupled to generate high flux in NADH.
    # Make them irreversible (reactions go in the direction of producing NH4).
    model.reactions.LEUDHr.lower_bound = 0
    model.reactions.VALDHr.lower_bound = 0
    model.reactions.ILEDHr.lower_bound = 0
    
    # PDHcr: dhlam_c + nad_c <=> h_c + lpam_c + nadh_c
    # iMO1053: forward direction
    model.reactions.PDHcr.lower_bound = 0
    
    
    return model
```

# Vary carbon:nitrogen ratio

```python
model = cobra.io.load_json_model('../find_protein_orthologues_PPKT2440_vs_PA/pruned_SI3_iJN1411final_flux_w_rhamnolipid_biosynthesis.json')
model = set_cplex_parameters(model, 0)
model = set_glycerol_minimal_medium(model, 0, 1, 1000)
#model = fix_NADH_NADPH_flux_loop(model)
model.reactions.ATPM.lower_bound = 0 # ignore maintenance flux

exchange_rxns_for_fva = ['EX_haa_e_', 'EX_lrhh_e_', 'EX_llrhh_e_', 
                         'EX_cit_e_', 'EX_icit_e_', 'EX_fum_e_', 'EX_succ_e_', 'EX_akg_e_', 'EX_mal_D_e_', 'EX_mal_L_e_', 'EX_acon_C_e_',
                         'EX_pyr_e_', 'EX_ac_e_',   'EX_lac_L_e_', 'EX_lac_D_e_', 'EX_etoh_e_', 'EX_co2_e_']
CN_ratio = list(np.arange(0,5,1))+list(np.arange(5,7,0.1))+list(np.arange(7,15,0.5))
exchange_flux_max = pd.DataFrame()
exchange_flux_min = pd.DataFrame()

for index,ratio in enumerate(CN_ratio):
    model.reactions.get_by_id('EX_glyc_e_').lower_bound = -ratio/3
    fva = cobra.flux_analysis.flux_variability_analysis(model, 
                                                        [r for r in model.reactions if r.id in exchange_rxns_for_fva],
                                                        loopless=True, 
                                                        fraction_of_optimum=0.999)
    
    if index==0:
        exchange_flux_max = fva['maximum'].to_frame()
        exchange_flux_min = fva['minimum'].to_frame()
    else:
        exchange_flux_max = pd.merge(exchange_flux_max, fva['maximum'].to_frame(), left_index=True, right_index=True)
        exchange_flux_min = pd.merge(exchange_flux_min, fva['minimum'].to_frame(), left_index=True, right_index=True)
```

```python
vary_CN_ratio_results = [CN_ratio, exchange_flux_min, exchange_flux_max] 
pickle.dump(vary_CN_ratio_results, open( "vary_CN_ratio_results.p", "wb" ) )
```

# Vary NADH flux in glycerol (carbon limited, C:N=3 and nitrogen limited, C:N=10)

```python
model = cobra.io.load_json_model('../find_protein_orthologues_PPKT2440_vs_PA/pruned_SI3_iJN1411final_flux_w_rhamnolipid_biosynthesis.json')
model = set_glycerol_minimal_medium(model, 10, 3, 1000) # nitrogen is limiting
model = set_cplex_parameters(model, 0)
model = fix_NADH_NADPH_flux_loop(model)
compound_id = 'nadh_c'
model, S_dict = add_flux_constraint(model, compound_id)

exchange_flux_max = pd.DataFrame()
exchange_flux_min = pd.DataFrame()
max_biomass_flux = []

nadh_flux = list(10**np.arange(-2,3.1,0.02))
for index,f in enumerate(nadh_flux):
    
    # set nadh flux constraint
    model.constraints["flux_constraint__%s"%(compound_id)].ub = f
    model.constraints["flux_constraint__%s"%(compound_id)].lb = f

    # optimization
    max_growth = model.slim_optimize()
    if str(max_growth) == 'nan':
        break
    max_biomass_flux.append(max_growth)
        
    # flux variability analysis
    exchange_rxns_for_fva = ['EX_haa_e_', 'EX_lrhh_e_', 'EX_llrhh_e_',  'EX_cit_e_', 'EX_icit_e_', 'EX_pyr_e_', 'EX_ac_e_',   'EX_lac_D_e_', 'EX_akg_e_']
    fva = cobra.flux_analysis.flux_variability_analysis(model, [r for r in model.reactions if r.id in exchange_rxns_for_fva], loopless=True, fraction_of_optimum=1)
    if index==0:
        exchange_flux_max = fva['maximum'].to_frame()
        exchange_flux_min = fva['minimum'].to_frame()
    else:
        exchange_flux_max = pd.merge(exchange_flux_max, fva['maximum'].to_frame(), left_index=True, right_index=True)
        exchange_flux_min = pd.merge(exchange_flux_min, fva['minimum'].to_frame(), left_index=True, right_index=True)
        
nadh_flux = [_f for _f in nadh_flux if _f < f]
```

```python
vary_nadh_results_CN10 = [nadh_flux, max_biomass_flux, exchange_flux_min, exchange_flux_max] 
pickle.dump(vary_nadh_results_CN10, open( "vary_NADH_flux_CNratio10_results.p", "wb" ) )
```

```python
model = cobra.io.load_json_model('../find_protein_orthologues_PPKT2440_vs_PA/pruned_SI3_iJN1411final_flux_w_rhamnolipid_biosynthesis.json')
compound_id = 'nadph_c'
model = set_glycerol_minimal_medium(model, 10, 10, 1000) # nitrogen is limiting
model = set_cplex_parameters(model, 0, compound_id)
model = fix_NADH_NADPH_flux_loop(model)
model, S_dict = add_flux_constraint(model, compound_id)

exchange_flux_max = pd.DataFrame()
exchange_flux_min = pd.DataFrame()
max_biomass_flux = []

nadph_flux = list(10**np.arange(0,3.05,0.02))
for index,f in enumerate(nadph_flux):
    
    # set nadh flux constraint
    model.constraints["flux_constraint__%s"%(compound_id)].ub = f
    model.constraints["flux_constraint__%s"%(compound_id)].lb = f

    # optimization
    model.solver.configuration.timeout = 1e10
    max_growth = model.slim_optimize()
    if str(max_growth) == 'nan':
        break
    max_biomass_flux.append(max_growth)
    print(f, max_growth)
    
    # flux variability analysis
    exchange_rxns_for_fva = ['EX_haa_e_', 'EX_lrhh_e_', 'EX_llrhh_e_',  'EX_cit_e_', 'EX_icit_e_', 'EX_pyr_e_', 'EX_ac_e_',   'EX_lac_D_e_', 'EX_akg_e_']
    model.solver.configuration.timeout = 30
    fva = cobra.flux_analysis.flux_variability_analysis(model, [r for r in model.reactions if r.id in exchange_rxns_for_fva], loopless=True, fraction_of_optimum=1.0)
    if index==0:
        exchange_flux_max = fva['maximum'].to_frame()
        exchange_flux_min = fva['minimum'].to_frame()
    else:
        exchange_flux_max = pd.merge(exchange_flux_max, fva['maximum'].to_frame(), left_index=True, right_index=True)
        exchange_flux_min = pd.merge(exchange_flux_min, fva['minimum'].to_frame(), left_index=True, right_index=True)
        
nadph_flux = [_f for _f in nadph_flux if _f < f]
```

```python
def multiple_testing_correction(pvalues, correction_type):
    pvalues = array(pvalues)
    sample_size = pvalues.shape[0]
    qvalues = empty(sample_size)
    if correction_type == "Bonferroni":
        # Bonferroni correction
        qvalues = sample_size * pvalues
    elif correction_type == "Bonferroni-Holm":
        # Bonferroni-Holm correction
        values = [(pvalue, i) for i, pvalue in enumerate(pvalues)]
        values.sort()
        for rank, vals in enumerate(values):
            pvalue, i = vals
            qvalues[i] = (sample_size-rank) * pvalue
    elif correction_type == "FDR":
        # Benjamini-Hochberg, AKA - FDR test
        values = [(pvalue, i) for i, pvalue in enumerate(pvalues)]
        values.sort()
        values.reverse()
        new_values = []
        for i, vals in enumerate(values):
            rank = sample_size - i
            pvalue, index = vals
            new_values.append((sample_size/rank) * pvalue)
        for i in range(0, int(sample_size)-1):
            if new_values[i] < new_values[i+1]:
                new_values[i+1] = new_values[i]
        for i, vals in enumerate(values):
            pvalue, index = vals
            qvalues[index] = new_values[i]
    return qvalues
```

```python
res = []
for rid in flux_samples_wt.columns:
    rxn = model_wt.reactions.get_by_id(rid)
    p1 = flux_samples_wt[rid]
    p2 = flux_samples_rhlA[rid]
    p1_mean = p1.mean()
    p2_mean = p2.mean()
    stat, pval = stats.ttest_ind(p1,p2, equal_var=False)
    
    # remove too large flux (seems to be in loop)
    if abs(p1_mean)>1e3 or abs(p2_mean)>1e3:
        continue
    res.append({'ID':rxn.id, 
                'Reaction':rxn.reaction,
                'Subsystem':rxn.subsystem,
                'WT_force_RL_mean':p1_mean,
                'RhlA_mean':p2_mean,
                'AbsDiff': abs(p2_mean-p1_mean),
                'Diff': p2_mean-p1_mean,
                'RelDiff': (p2_mean-p1_mean)/p1_mean,
                'pval':pval})    
df_flux_diff = pd.DataFrame(res)
df_flux_diff['pval'] = multiple_testing_correction(df_flux_diff['pval'], correction_type="FDR")
df_flux_diff.set_index('Reaction',inplace=True)
df_flux_diff['-logpval'] = -np.log(df_flux_diff['pval'])
```

```python
df_flux_diff_filtered = df_flux_diff[(df_flux_diff['pval'] <= 0.05) & (df_flux_diff['AbsDiff']>=1e-6)]
df_flux_diff_filtered.head()
```

```python
df_kb_rep1_neg = pd.read_excel('../data/metabolomics/metabolomics_KerryBoyle.xlsx',index_col=0, sheet_name='Replicate 1 negative mode')
df_kb_rep1_pos = pd.read_excel('../data/metabolomics/metabolomics_KerryBoyle.xlsx',index_col=0, sheet_name='Replicate 1 positive mode')
df_kb_rep2_neg = pd.read_excel('../data/metabolomics/metabolomics_KerryBoyle.xlsx',index_col=0, sheet_name='Replicate 2 negative mode')
df_kb_rep2_pos = pd.read_excel('../data/metabolomics/metabolomics_KerryBoyle.xlsx',index_col=0, sheet_name='Replicate 2 positive mode')
df_kb_rep1_neg = df_kb_rep1_neg.iloc[:,[4,5,6,16,17,18]]
df_kb_rep1_pos = df_kb_rep1_pos.iloc[:,[4,5,6,16,17,18]]
df_kb_rep2_neg = df_kb_rep2_neg.iloc[:,[4,5,6,16,17,18]]
df_kb_rep2_pos = df_kb_rep2_pos.iloc[:,[4,5,6,16,17,18]]
df_kb_rep1 = pd.concat([df_kb_rep1_neg,df_kb_rep1_pos])
df_kb_rep2 = pd.concat([df_kb_rep2_neg,df_kb_rep2_pos])
df_kb = pd.concat([df_kb_rep1, df_kb_rep2], axis=1)
df_kb['WT_mean'] = np.nanmean(df_kb[[colname for colname in df_kb.columns if 'WT' in colname]], axis=1)
df_kb['RhlA_mean'] = np.nanmean(df_kb[[colname for colname in df_kb.columns if 'rhlA' in colname]], axis=1)
df_kb = pd.merge(df_kb, df_KEGG_mapping, left_index=True, right_index=True, how='outer')
# some metabolites are not reproducible
df_kb = df_kb.drop('3 4-Dihydroxy-2-butanone 4-phosphate', axis=0)
df_kb.head()
```

```python
lines2append = []
for cpd in df_kb.index:
    log2FC = np.log2(df_kb.loc[cpd,'RhlA_mean']/df_kb.loc[cpd,'WT_mean'])
    p1 = list(df_kb.loc[cpd,[x for x in df_kb.columns if 'WT' in x]])
    p2 = list(df_kb.loc[cpd,[x for x in df_kb.columns if 'rhlA' in x]])
    p1 = [x for x in p1 if str(x) != 'nan']
    p2 = [x for x in p2 if str(x) != 'nan']
    if len(p1)>0 and len(p2)>0:
        stat, pval = stats.ttest_ind(p1, p2, equal_var=False)
        lines2append.append([cpd, df_kb.loc[cpd,'KEGG compound'], log2FC, pval])
df_kb_summary = pd.DataFrame(lines2append, columns=['Name','KEGG ID','log2FC','pval'])
df_kb_summary['pval'] = multiple_testing_correction(df_kb_summary['pval'], correction_type="FDR")
df_kb_summary['-logpval'] = -np.log(df_kb_summary['pval'])
df_kb_summary.head()
```

## pathway level summary

```python
df_kegg_pathway = pd.read_csv('kegg_pathway.txt', sep='\t', header=None)
df_kegg_pathway.columns = ['KEGG pathway','KEGG compound']
df_kegg_pathway['KEGG pathway'] = [x.lstrip('path:') for x in df_kegg_pathway['KEGG pathway']]
df_kegg_pathway['KEGG compound'] = [x.lstrip('cpd:') for x in df_kegg_pathway['KEGG compound']]
df_kegg_pathway.head()
```

```python
unique_kegg_pathway = np.unique(df_kegg_pathway['KEGG pathway'])
lines2append = []
for curr_pathway in unique_kegg_pathway:
    curr_compounds = list(df_kegg_pathway.loc[df_kegg_pathway['KEGG pathway']==curr_pathway,'KEGG compound'])
    curr_df_kb = df_kb[df_kb['KEGG compound'].isin(curr_compounds)]
    for cpd in curr_df_kb.index:
        log2FC = np.log2(curr_df_kb.loc[cpd,'RhlA_mean']/curr_df_kb.loc[cpd,'WT_mean'])
        p1 = list(curr_df_kb.loc[cpd,[x for x in curr_df_kb.columns if 'WT' in x]])
        p2 = list(curr_df_kb.loc[cpd,[x for x in curr_df_kb.columns if 'rhlA' in x]])
        p1 = [x for x in p1 if str(x) != 'nan']
        p2 = [x for x in p2 if str(x) != 'nan']
        if len(p1)>0 and len(p2)>0:
            stat, pval = stats.ttest_ind(p1, p2, equal_var=False)
            lines2append.append([curr_pathway, curr_df_kb.loc[cpd,'KEGG compound'], log2FC, pval])
df_pathway_summary = pd.DataFrame(lines2append, columns=['KEGG pathway','KEGG compound','log2FC','pval'])
df_pathway_summary['pval'] = multiple_testing_correction(df_pathway_summary['pval'], correction_type="FDR")
df_pathway_summary['-logpval'] = -np.log(df_pathway_summary['pval'])
df_pathway_summary.head()
```
