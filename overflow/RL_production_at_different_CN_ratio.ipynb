```python
import cobra
import numpy as np
from cobra import Model, Reaction, Metabolite
import matplotlib.pyplot as plt
import pandas as pd
from cobra.flux_analysis import flux_variability_analysis
```

# Function to set non-carbon nutrients

```python
def set_non_carbon_nutrient_flux(model):
    
    # set lower bound of all exchange reactions to zero
    for rxn in model.exchanges:
        rxn.lower_bound = 0

    # turn on specific fluxes
    minimal__media = {
            'EX_nh4_e_': -1,         # ammonium
            'EX_pi_e_':  -1000,      # phosphate
            'EX_so4_e_': -1000,      # sulfate
            'EX_fe3_e_': -1000,      # fe3+
            'EX_mn2_e_': -1000,      # mn2+
            'EX_zn2_e_': -1000,      # zn2+
            'EX_cu2_e_': -1000,      # cu2+
            'EX_ca2_e_': -1000,      # ca2+
            'EX_cl_e_' : -1000,      # cl-
            'EX_cobalt2_e_': -1000,  # cobalt
            'EX_k_e_':   -1000,      # k
            'EX_mobd_e_'   : -1000,  # mo6+
            'EX_na1_e_'    : -1000,     # na+
            'EX_ni2_e_'    : -1000,     # ni2+
            'EX_mg2_e_': -1000,      # mg2+
            'EX_o2_e_':  -1000
        }

    for r in minimal__media.keys():
        try:
            rxn = model.reactions.get_by_id(r)
            rxn.lower_bound = minimal__media[r]
            rxn.upper_bound = 1000
        except:
            print("The reaction {} does not exist in model.".format(str(r)))
            continue
            
    return model
```

# Pseudomonas putida model: HAA flux at different C:N ratios

```python
carbon_sources = ['glycerol','glucose','succinate','acetate']
exchange_rxns = ['EX_glyc_e_', 'EX_glc_e_', 'EX_succ_e_', 'EX_ac_e_'] # exchange reactions of glycerol, glucose, succinaate and acetate
carbon_atom_numbers = [3, 6, 4, 2] # number of carbon atoms in each source

CN_ratio = np.arange(0.55,20.05,0.05)
max_HAA_flux = {}
min_HAA_flux = {}
max_biomass_flux = {}

for index,substrate in enumerate(carbon_sources):
    model = cobra.io.load_json_model('../add_RL_production/SI3_iJN1411final_flux_w_rhamnolipid_biosynthesis.json')
    model.solver = 'cplex'
    model = set_non_carbon_nutrient_flux(model)

    max_HAA_flux_for_current_carbon_source = []
    min_HAA_flux_for_current_carbon_source = []
    max_biomass_flux_for_current_carbon_source = []
    for ratio in CN_ratio:
        model.reactions.get_by_id(exchange_rxns[index]).lower_bound = -ratio/carbon_atom_numbers[index] # keep in mind that nitrogen (NH4) flux is 1
        model.reactions.get_by_id(exchange_rxns[index]).upper_bound = -ratio/carbon_atom_numbers[index] # 
        max_biomass_flux_for_current_carbon_source.append(model.slim_optimize()) # maximize growth flux
        fva = cobra.flux_analysis.flux_variability_analysis(model, model.reactions.EX_haa_e_, loopless=True, fraction_of_optimum=1.00) # find flux range of HAA reaction
        min_HAA_flux_for_current_carbon_source.append(fva.loc['EX_haa_e_','minimum'])
        max_HAA_flux_for_current_carbon_source.append(fva.loc['EX_haa_e_','maximum'])
        
    min_HAA_flux[substrate] = min_HAA_flux_for_current_carbon_source
    max_HAA_flux[substrate] = max_HAA_flux_for_current_carbon_source
    max_biomass_flux[substrate] = max_biomass_flux_for_current_carbon_source
```

## Plot max HAA flux vs. C:N ratio

```python
plt.figure(figsize=(8, 8))
colors = ['black','blue','green','red']
for index,substrate in enumerate(carbon_sources):
    #plt.plot(CN_ratio, min_HAA_flux[substrate], label=carbon_sources[index], color=colors[index], linewidth=2)  
    plt.plot(CN_ratio, max_HAA_flux[substrate], label=carbon_sources[index], color=colors[index], linewidth=2)
plt.legend()
plt.xlabel('C:N ratio', fontsize=12)
plt.ylabel('Maximum HAA flux (mmol/gDW/h)', fontsize=12)
plt.xlim([0,20])
plt.show()
```

## Plot max biomass flux vs. C:N ratio

```python
plt.figure(figsize=(8, 8))
colors = ['black','blue','green','red']
for index,substrate in enumerate(carbon_sources):
    plt.plot(CN_ratio, max_biomass_flux[substrate], label=carbon_sources[index], color=colors[index], linewidth=2)  
plt.legend()
plt.xlabel('C:N ratio', fontsize=12)
plt.ylabel('Maximum biomass flux (mmol/gDW/h)', fontsize=12)
plt.xlim([0,20])
plt.ylim([0,0.1])
plt.show()
```
